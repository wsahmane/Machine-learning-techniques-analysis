# Classification

```python
psyco_numerical.keys()
```

```python
rearrange_list = ['age', 'occupation', 'line_of_work', 'time_bp', 'time_dp',
       'travel_time', 'easeof_online', 'home_env', 'prod_inc', 'sleep_bal',
       'new_skill', 'fam_connect', 'relaxed', 'self_time', 'like_hw',
       'dislike_hw', 'prefer', 'certaindays_hw', 'gender']

psyco_clsf = psyco_numerical[rearrange_list]
```

```python
psyco_clsf.head()
```

```python
psyco_clsf['gender'].value_counts()
```

```python
psyco_clsf['gender'].replace({2: 0}, inplace=True)
```

```python
psyco_clsf['gender'].value_counts()
```

```python
# Predictor
X = psyco_clsf.iloc[:,:-1].values

# Target attribute - gender
y = psyco_clsf.iloc[:,-1].values
```

```python
psyco_clsf.iloc[:,:-1].values
```

```python
psyco_clsf.iloc[:,-1].values
```

```python
# Training and testing samples
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state =0)
```

```python
# Decision Tree 

from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier(criterion='entropy', random_state=0)

#fitting the data into the model
DT.fit(X_train,y_train)
```

```python
y_pred = DT.predict(X_test)
y_pred
```

```python
from sklearn import tree, metrics

# Evaluation measures
print ("Accuracy = %.2f" % metrics.accuracy_score(y_test, y_pred))
print ("Precision: %.2f" % metrics.precision_score(y_test, y_pred))
print ("Recall: %.2f" % metrics.recall_score(y_test, y_pred))
print ("F1 Measure: %.2f" % metrics.f1_score(y_test, y_pred))
print ("Confusion Matrix:")
metrics.confusion_matrix(y_test, y_pred)
```

```python
#cross-validation 
from sklearn.model_selection import cross_val_score
DT = DecisionTreeClassifier(criterion='entropy', random_state=0)

#accuracy scores; increase cv to 20 get better Accuracy 
scores = cross_val_score(DT, X, y, cv=20)
print(scores)


# Average accuracy accross 
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

```python
#performance metrics

from sklearn.model_selection import cross_val_predict
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

#predicted values
y_pred = cross_val_predict(DT, X, y, cv=20)

print(confusion_matrix(y, y_pred), '\n')
print('Mean Absolute Error:', metrics.mean_absolute_error(y, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, y_pred)), '\n')

print(classification_report(y, y_pred))
print("Accuracy:", accuracy_score(y, y_pred)) 
#print("AUC:", roc_auc_score(y_test, y_pred))
```

```python
# Naive Bayes 

from sklearn.naive_bayes import GaussianNB


# initializing Naive Bayes 
NB = GaussianNB()

# fitting the classifier to the data
NB.fit(X_train, y_train)

# using the resulting model to predict value on the test set
y_pred = NB.predict(X_test)

print(confusion_matrix(y_test, y_pred), '\n')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)), '\n')

print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# cross-validation 
NB1 = GaussianNB()

# accuracy scores 
scores = cross_val_score(NB1, X, y, cv=20)
print(scores)

# printing the average 
print("Accuracy: %0.2f(+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

```python
# performance metrics 

from sklearn.model_selection import cross_val_predict

# predicted values
y_pred = cross_val_predict(NB1, X, y, cv=20)

print(confusion_matrix(y, y_pred), '\n')
print('Mean Absolute Error:', metrics.mean_absolute_error(y, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, y_pred)), '\n')

print(classification_report(y, y_pred))
print("Accuracy:", accuracy_score(y, y_pred))
```

```python
#k-Nearest Neighbour    


from sklearn.neighbors import KNeighborsClassifier

# default values applied: minkowski = distance formula & p = 2 gives Eucledian distance
# 5 neighbours selected

knn = KNeighborsClassifier(n_neighbors = 5)

# fitting the classifier to the data
knn.fit(X_train, y_train)

# predicting values using the knn model
y_pred = knn.predict(X_test)

# performance metrics

print(confusion_matrix(y_test, y_pred), '\n')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)), '\n')

print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#Linear Regression

from sklearn.linear_model import LinearRegression

# initialize 
LR = LinearRegression()

#fit the classifier to the data
LR.fit(X_train,y_train)

#use the resulting model to predict value on the test sey
y_pred = LR.predict(X_test)
```
